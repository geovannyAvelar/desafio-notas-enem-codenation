```python
import pandas as pd
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import RobustScaler
from sklearn.neural_network import MLPRegressor
```

```python
train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')
```

```python
df = train[['NU_NOTA_CN', 'NU_NOTA_CH', 'NU_NOTA_LC', 'NU_NOTA_MT', 'NU_NOTA_REDACAO']]
```

```python
df['NU_NOTA_CN'] = df['NU_NOTA_CN'].fillna(df['NU_NOTA_CN'].mean())
df['NU_NOTA_CH'] = df['NU_NOTA_CH'].fillna(df['NU_NOTA_CH'].mean())
df['NU_NOTA_LC'] = df['NU_NOTA_LC'].fillna(df['NU_NOTA_LC'].mean())
df['NU_NOTA_MT'] = df['NU_NOTA_MT'].fillna(df['NU_NOTA_MT'].mean())
df['NU_NOTA_REDACAO'] = df['NU_NOTA_REDACAO'].fillna(df['NU_NOTA_REDACAO'].mean())
```

```python
X = df[['NU_NOTA_CN', 'NU_NOTA_CH', 'NU_NOTA_LC', 'NU_NOTA_REDACAO']]
y = df['NU_NOTA_MT']
```

```python
scaler = RobustScaler()
scaled_x = scaler.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(scaled_x, y, test_size=0.33, random_state=532543)
```

```python
mlp = MLPRegressor(solver='lbfgs', hidden_layer_sizes=(5, 5), activation = 'relu', random_state=1432, max_iter = 400)
mlp.fit(X_train, y_train)
mlp.score(X_test, y_test) * 100
```
